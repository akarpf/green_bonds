- Settlement Date: shows the date on which the transaction is completed, which is when final payment for the bonds and transfer of ownership of the bond occurs. Transactions in the secondary market typically settle three business days after the trade date – this is typically referred to as a "regular way" trade with a T+3 settlement. ##### 2 flavours of Munis
Municipal bonds come in the following two varieties:

- general obligation bonds (GO)
- revenue bonds

General obligation bonds, issued to raise immediate capital to cover expenses, are supported by the taxing power of the issuer. Revenue bonds, which are issued to fund infrastructure projects, are supported by the income generated by those projects. Both types of bonds are tax exempt and particularly attractive to risk-averse investors due to the high likelihood that the issuers will repay their debts.

##### Risk Factors 
While buying municipals bonds is viewed as a conservative investment strategy, it is not risk-free. The following are the risk factors:

- Credit Risk: If the issuer is unable to meets its financial obligations, it may fail to make scheduled interest payments and/or be unable to repay the principal upon maturity. To assist in the evaluation of an issuer's creditworthiness, ratings agencies (such as Moody's Investors Service and Standard & Poor's), analyze a bond issuer's ability to meet its debt obligations, and issue ratings from 'Aaa' or 'AAA' for the most creditworthy issuers to 'Ca', 'C', 'D', 'DDD', 'DD' or 'D' for those in default. Bonds rated 'BBB', 'Baa' or better are generally considered appropriate investments when capital preservation is the primary objective. To reduce investor concern, many municipal bonds are backed by insurance policies guaranteeing repayment in the event of default.

- Interest-Rate Risk: The interest rate of most municipal bonds is paid at a fixed rate. The rate does not change over the life of the bond. If interest rates in the marketplace rise, the bond you own will be paying a lower yield relative to the yield offered by newly issued bonds.

- Tax-Bracket Changes: Municipal bonds generate tax-free income, and therefore pay lower interest rates than taxable bonds. Investors who anticipate a significant drop in their marginal income-tax rate may be better served by the higher yield available from taxable bonds.

- Call Risk: Many bonds allow the issuer to repay all or a portion of the bond prior to the maturity date. The investor's capital is returned with a premium added in exchange for the early debt retirement. While you get your entire initial investment plus some back if the bond is called, your income stream ends earlier than you were expecting it to.

- Market Risk: The underlying price of a particular bond changes in response to market conditions. When interest rates fall, newly issued bonds will pay a lower yield than existing issues, which makes the older bonds more attractive. Investors who want the higher yield may be willing to pay a premium to get it. Likewise, if interest rates rise, newly issued bonds will pay a higher yield than existing issues. Investors who buy the older issues are likely to do so only if they get it at a discount. If you buy a bond and hold it until maturity, market risk is not a factor because your principal investment will be returned in full at maturity. Should you choose to sell prior to the maturity date, your gain or loss will be dictated by market conditions, and the appropriate tax consequences for capital gains or losses will apply.

- Initial offering Price – The price at which a new issue of municipal securities is offered to the public at the time of original issuance. This price is sometimes referred to as the “public offering price.” The initial offering price may be at a discount to par, par or a premium to par.


#### Term Structure of Interest Rates

##### Bond Price & Yields (Basics)

$$
P=\frac{R}{(1+rs_1)} + \frac{R}{(1+rs_2)^2} + \ldots + \frac{R + RV}{(1+rs_n)^n}
$$

$RV$ ... is the redemption payment on maturity (par)

$R$ ... Coupon

$rs$ ... spot interest rate applicable to the cash flow in period $t = 1 \ldots n$   $(rs_t = (M + C)/P_t)$

The bond price equation can be given in terms of discount factors, i.e. the present value of each coupon payment and the maturity payment being the product of multiplying them by their relevant discount factors. Let $df_t$ denote the $t$-period discount factor $t=1,\ldots,m$:

$$
df_t=\frac{1}{(1+rs)^t} = (1+rs)^{-t}, \quad t=1,\ldots,m
$$ 

With this the bond price equation can be reformulated to:

$$
P=\sum_{t=1}^{n} Rdf_t + RVdf_n
$$

The spot rate $rs$ is the average of the implied short-term forward rates $rf_1, rf_2, \ldots, rf_t$. It can thus be stated:

$$
1/df_1 = (1 + rs_1) = (1 + rf_1)\\
1/df_2 = (1 + rs_2)^2 = (1 + rf_1)(1 + rf_2)\\
\ldots \\
1/df_t = (1 + rs_t)^t = (1 + rf_1)(1 + rf_2)\ldots(1 + rf_t)
$$

$1 + rs_t$ is thus the geometric mean of $(1 + rf_1)(1 + rf_2)\ldots(1 + rf_t)$. Implied forward rates indicate the expected short-term (one-period) future interest rate for a specific point along the term structure; they reflect the spread on the marginal rate of return that the market requires if it is investing in debt instruments of longer and longer maturities.

##### Bond Price Calculations along MSRB norms

###### maturity longer than period between interest payments 

$$
P = \left[\frac{RV}{\left(1 + \frac{Y}{M}\right)^{N - 1 + \frac{E - A}{E}} }\right] +
\left[ \sum_{K=1}^{N} \frac{ 100 \cdot \frac{R}{M}}{\left(1 + \frac{Y}{M}\right)^{K - 1 + \frac{E - A}{E}} }\right] - 
\left[100 \cdot \frac{A}{B} \cdot R \right]
$$ 

###### maturity hosrter than period between interest payments 

$$
P = \left[\frac{\frac{RV}{100} \cdot \frac{R}{M}}{ 1 + \left(\frac{E - A}{E} \cdot \frac{Y}{M} \right)}\right] -
\left[ \frac{A}{B} \cdot R \right]
$$


###### Variables: price and yield computation along: http://www.msrb.org/Rules-and-Interpretations/MSRB-Rules/General/Rule-G-33.aspx

- A is the number of accrued days from the beginning of the interest payment period to the settlement date (computed in accordance with the provisions of section (e) below);

- B is the number of days in the year (computed in accordance with the provisions of section (e) below);

- E is the number of days in the interest payment period in which the settlement date falls (computed in accordance with the provisions of section (e) below);

- M is the number of interest payment periods per year standard for the security involved in the transaction;

- P is the dollar price of the security for each $100 par value (divided by 100);

- R is the annual interest rate (expressed as a decimal);

- RV is the redemption value of the security per $100 par value; and

- Y is the yield price of the transaction (expressed as a decimal).

###### Day Counting.

Day Count Basis: Computations under the requirements of this rule shall be made on the basis of a thirty-day month and a three-hundred-sixty-day year, or, in the case of computations on securities paying interest solely at redemption, on the day count basis selected by the issuer of the securities.

Day Count Formula: For purposes of this rule, computations of day counts on the basis of a thirty-day month and a three-hundred-sixty-day year shall be made in accordance with the following formula.

Number of Days = $(Y2 - Y1) 360 + (M2 - M1) 30 + (D2 - D1)$

For purposes of this formula the symbols shall be defined as follows:

- M1 is the month of the date on which the computation period begins;

- D1 is the day of the date on which the computation period begins;

- Y1 is the year of the date on which the computation period begins;

- M2 is the month of the date on which the computation period ends;

- D2 is the day of the date on which the computation period ends; and

- Y2 is the year of the date on which the computation period ends.

For purposes of this formula, if the symbol $D2$ has a value of $3$," and the symbol $D1$ has a value of $30$ or $31$," the value of the symbol $D2$ shall be changed to $30$. If the symbol $D1$ has a value of $31$, the value of the symbol $D1$ shall be changed to $30$. For purposes of this rule time periods shall be computed to include the day specified in the rule for the beginning of the period but not to include the day specified for the end of the period.

```python
#count the days along the MSRB standards
def daycountmuni(start,end):
    #If the symbol "D2" has a value of "31," and the symbol "D1" has a value of "30" or "31," the value of the symbol "D2" shall be changed to "30." 
    #If the symbol "D1" has a value of "31," the value of the symbol "D1" shall be changed to "30."
    d1year = start.year
    d1month = start.month
    d1day = start.day
    d2year = end.year
    d2month = end.month
    d2day = end.day
    if d1day == 31:
        d1day = 30;
    if d2day == 31 and d1day == 30:
        d2day = 30;
    numbofdays = 360 * (d2year - d1year) + 30 * (d2month - d1month) + (d2day - d1day)
    return numbofdays

#yield to maturity
def ytm(settlement,maturity,price,coupon,face = 100,freq=2):
    p = price
    r = coupon/100
    m = freq
    rv = face
    remaining = daycountmuni(settlement,maturity)
    ipd = [maturity - relativedelta(months=+int(12/m)*i) for i in range(0,500) if settlement < (maturity - relativedelta(months=+int(12/m)*i))]
    n = len(ipd)
    nt = [i+1 for i in range(int(n))]
    a = daycountmuni(min(ipd) - relativedelta(months=+int(12/m)),settlement)
    b = 360
    e = 360/m
    ytmfuncs = lambda y: (rv/100 + r/m)/(1 + (e - a)/e * (y/m)) - (r * a/b)                                      
    ytmfuncl = lambda y: (rv/((1+y/m)**(n-1+(e-a)/e))) + sum([((100*r/m)/((1+y/m)**(k-1+(e-a)/e))) for k in nt]) - 100*(a/b)*r - p
    if remaining < e:
        result = optimize.newton(ytmfuncs, r)*100
        return float(result)
    else:
        result = optimize.newton(ytmfuncl, r)*100
        return float(result)

def getttc(p, r, yld, settlement, maturity, face = 100, m = 2):
    def obfun(ttc, p = p, r = r, yld = yld, settlement = settlement, face = face, m = m):
        y = yld/100 
        r = r/100
        rv = face
        calldate = settlement + relativedelta(days=+int(ttc))
        pers = math.ceil(ttc/(12/m * 30))
        ipd = []
        for i in range(0,pers):
            cpdate = calldate - relativedelta(months=+int(12/m)*i)
            if settlement < cpdate: 
                ipd.append(cpdate)
        
        n = len(ipd)
        nt = [i+1 for i in range(int(n))]
        a = daycountmuni(min(ipd) - relativedelta(months=+int(12/m)),settlement)
        b = 360
        e = 360/m
        if ttc < 360:
            res = (rv/100 + r/m)/(1 + (e - a)/e * (y/m)) - (r * a/b)
        else:    
            res = (rv/((1+y/m)**(n-1+(e-a)/e))) + sum([((100*r/m)/((1+y/m)**(k-1+(e-a)/e))) for k in nt]) - 100*(a/b)*r - p
        return res ** 2

    dic = {}
    remdays = daycountmuni(settlement,maturity)
    for i in range(1,remdays):
        dic[i] = obfun(ttc = i)
    ttc = min(dic,key=dic.get) + 1
    calldate = settlement + relativedelta(days=+int(ttc))
    return calldate

#get call date
def getcalldate(data):
    for index, row in data.iterrows():
        data.loc[index,'calldate'] = getttc(row['price'],row['coupon'],row['yld'],row['settlementdate'],row['maturitydate'],face = 100, m = 2)
    return data

def spr(settlement,maturity,price,coup,face):
    tenor = ql.Period(ql.Semiannual)
    calendar = ql.UnitedStates()
    bizConvention = ql.Unadjusted
    dateGeneration = ql.DateGeneration.Backward
    monthEnd = True
    schedule = ql.Schedule(toqtd(settlement),toqtd(maturity),tenor,calendar,bizConvention,bizConvention,dateGeneration,monthEnd)
    rmd = len(schedule)
    cou = coup/2
    yof = lambda r:  sum([coup/((1+r) ** t) for t in range(1,rmd-1)]) + (coup + face)/((1+r) ** rmd) - price
    return optimize.newton(yof,coup/100)

def splreg(dat):
    varstr = 'ttm'
    rsform = "bs(" + varstr + ",df = 6, degree=3, include_intercept=False) - 1"
    form = "yld ~ " + rsform
    dat = dat.sort_values(varstr)
    res = sm.ols(formula = form, data=dat).fit()
    lildat = pd.DataFrame(pa.dmatrix(rsform, data = dat), index = range(1, dat.shape[0] + 1))
    newstr = "prediction" + varstr
    newstr = np.array(np.dot(res.params.values, np.matrix(lildat).T).T)
    res = pd.DataFrame({'ttm' : dat['ttm'].values.tolist(), 'yld' : [i[0] for i in newstr.tolist()]})
    return res

def log_progress(sequence, every=None, size=None):
    from ipywidgets import IntProgress, HTML, VBox
    from IPython.display import display

    is_iterator = False
    if size is None:
        try:
            size = len(sequence)
        except TypeError:
            is_iterator = True
    if size is not None:
        if every is None:
            if size <= 200:
                every = 1
            else:
                every = int(size / 200)     # every 0.5%
    else:
        assert every is not None, 'sequence is iterator, set every'

    if is_iterator:
        progress = IntProgress(min=0, max=1, value=1)
        progress.bar_style = 'info'
    else:
        progress = IntProgress(min=0, max=size, value=0)
    label = HTML()
    box = VBox(children=[label, progress])
    display(box)

    index = 0
    try:
        for index, record in enumerate(sequence, 1):
            if index == 1 or index % every == 0:
                if is_iterator:
                    label.value = '{index} / ?'.format(index=index)
                else:
                    progress.value = index
                    label.value = u'{index} / {size}'.format(
                        index=index,
                        size=size
                    )
            yield record
    except:
        progress.bar_style = 'danger'
        raise
    else:
        progress.bar_style = 'success'
        progress.value = index
        label.value = str(index or '?')
        
def parallelize_dataframe(df, func):
    df_split = np.array_split(df,4)
    pool = Pool(4)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df

def findstate(name):
    m = re.search(r"\(([A-Z]+)\)",name)
    return m.group(1)
```

```python
def moodycv(item):
    if item in ['Aaa']: 
        return 1 #'AAA'
    if item in ['Aaa','Aa1','Aa2','Aa3']: 
        return 2 #'AA'
    if item in ['A1','A2','A3','A31','A32']: 
        return 3 #'A'
    if item in ['Baa1','Baa2','Baa3']: 
        return 4 #'BBB'
    if item in ['Ba1','Ba2','Ba3']: 
        return 5 #'BB'
    if item in ['B1','B2','B3']: 
        return 6 #'B'
    if item in ['Caa1','Caa2','Caa3']: 
        return 7 #'CCC'
    if item in ['Ca']: 
        return 8 #'CC'
    if item in ['C']: 
        return 10 #'D'
    if item == None or item == 'WR':
        return 'NaN'
    
def fitchcv(item):
    if item in ['AAA','AAANa']: 
        return 1 #'AAA' 
    if item in ['AA+','AA','AA-','AA+/A-1']: 
        return 2 #'AA' 
    if item in ['A+','A','A-']: 
        return 3 #'A'
    if item in ['BBB+','BBB','BBB-']: 
        return 4 #'BBB'
    if item in ['BB+','BB','BB-']: 
        return 5 #'BB'
    if item in ['B+','B','B-']: 
        return 6 #'B'
    if item in ['CCC']: 
        return 7 #'CCC'
    if item in ['DDD','DD','D']: 
        return 10 #'D'
    if item == None or item == 'NR' or item == 'WD':
        return 'NaN'
    
def spcv(item):
    if item in ['AAA']: 
        return 1 #'AAA'
    if item in ['AA+','AA','AA-','AA-1A-1 +','AA+/A-1','AANA-1+','AA+/A-1 +']: 
        return 2 #'AA'
    if item in ['A+','A','A-','A/A-1','ANA-1+','A+/A-1','ANA-1','A-1A-2','ANA-2']: 
        return 3 #'A'
    if item in ['BBB+','BBB','BBB-']: 
        return 4 #'BBB'
    if item in ['BB+','BB','BB-']: 
        return 5 #'BB'
    if item in ['B+','B','B-']: 
        return 6 #'B'
    if item in ['CCC+','CCC','CCC-']: 
        return 7 #'CCC'
    if item in ['CC']: 
        return 8 #'CC'
    if item in ['C']: 
        return 9 #'C'
    if item in ['D']: 
        return 10 #'D'
    if item == None or item == 'NR':
        return 'NaN'
    
def kbracv(item):
    if item in ['AAA+','AAA','AAA-']: 
        return 1 #'AAA'
    if item in ['AA+','AA','AA-']: 
        return 2 #'AA'
    if item in ['A+','A','A-']: 
        return 3 #'A'
    if item in ['BBB+','BBB','BBB-']: 
        return 4 #'BBB'
    if item in ['BB+','BB','BB-']: 
        return 5 #'BB'
    if item in ['B+','B','B-']: 
        return 6 #'B'
    if item in ['CCC+','CCC','CCC-']: 
        return 7 #'CCC'
    if item in ['CC+','CC','CC-']: 
        return 8 #'CC'
    if item in ['C+','C','C-']: 
        return 9 #'C'
    if item in ['D+','D','D-']: 
        return 10 #'D'
    if item == None or item == 'WR':
        return 'NaN'
    
def rating(item):
    if 1 <= item <= 1.5:
        return 'AAA'
    if 1.5 < item <= 2.5:
        return 'AA'
    if 2.5 < item <= 3.5:
        return 'A'
    if 3.5 < item <= 4.5:
        return 'BBB'
    if 4.5 < item <= 5.5:
        return 'BB'
    if 5.5 < item <= 6.5:
        return 'B'
    if 6.5 < item <= 7.5:
        return 'CCC'
    if 7.5 < item <= 8.5:
        return 'CC'
    if 8.5 < item <= 9.5:
        return 'C'
    if 9.5 < item <= 10.0:
        return 'D'
    if np.isnan(item):
        return 'NR'
```

```python
def toqtd(d):
    return ql.Date(d.day,d.month,d.year)

bzco = ql.Unadjusted
dcnt = ql.Thirty360()
freq = ql.Period(ql.Semiannual)
cale = ql.UnitedStates()
dategenerate = ql.DateGeneration.Backward
compounding = ql.Compounded
eom = True
sett = 3
face = 100

def ytc(data,idx):
    price = ql.SimpleQuote(data.price)
    tradedate = toqtd(data.tradedate)
    coupon = data.coupon/100
    issuedate = toqtd(data.dateddate)
    maturity = toqtd(data.maturitydate)
    calldate = toqtd(data.calldate)
    dtmr = int(maturity - tradedate)
    dtcr = int(calldate - tradedate)

    if (maturity - calldate) < 90:
        calldate = maturity
    
    ###
    ql.Settings.instance().evaluationDate = tradedate
    ###
    schedule = ql.Schedule(issuedate,maturity,freq,cale,bzco,bzco,dategenerate,eom)
    schedule2worst = ql.Schedule(issuedate,calldate,freq,cale,bzco,bzco,dategenerate,eom)
    pdate = np.array(list(schedule))
    calldates = pdate[pdate > calldate]
    callabilityschedule = ql.CallabilitySchedule()
    callprice = face
    for d in calldates:
        callabilityprice = ql.CallabilityPrice(callprice,ql.CallabilityPrice.Clean)
        callabilityschedule.append(ql.Callability(callabilityprice,ql.Callability.Call,d));
    ###
    try: 
        callbond = ql.CallableFixedRateBond(sett,face,schedule,[coupon],dcnt,compounding,face,issuedate,callabilityschedule)
        callytm = callbond.bondYield(data.price,dcnt,compounding,ql.Semiannual)
    except:
        callytm = None
    try:
        fixbond = ql.FixedRateBond(sett,face,schedule2worst,[coupon],dcnt,compounding,face)
        fixytm = fixbond.bondYield(data.price,dcnt,compounding,ql.Semiannual)
    except:
        fixytm = None
        
    return (idx,callytm,fixytm,dtmr,dtcr)

def getytc(data):
    out = []
    pbar = tqdm(data,miniters = 1000)
    for i, row in data.iterrows():
        out.append(ytc(row,int(i)))
        pbar.update(1)
    pbar.close()    
    return out

#from multiprocessing import Pool
def paradf(data, func):
    dfsplit = np.array_split(data,4)
    pool = Pool(4)
    df = pool.map_async(func,dfsplit)
    pool.close()
    pool.join()
    res = df.get()
    return res
```

```python
tradda = pd.read_pickle("traddisagg.p")

tradda = tradda[(tradda.ytm > 0) & (tradda.ytm < np.percentile(tradda.ytm,99.9))]
tradda = tradda[(tradda.ytc > 0) & (tradda.ytc < np.percentile(tradda.ytc,99.9))]

def yldcurvedata(data,today = ql.Date(23,1,2017)):
    data = data.sort_values(by ='dtm', ascending = False)
    data = data.groupby(['dtm'])
    data = data.agg({'ytm':'mean','ytc':'mean','coupon':'mean','price':'mean','amount':'sum'})
    data = data.reset_index()
    
    ql.Settings.instance().evaluationDate = today
    bond_helpers = []
    
    for i, row in data.iterrows():
        maturity = today + int(row.dtm)
        schedule = ql.Schedule(today,maturity,freq,cale,bzco,bzco,dategenerate,eom)
        price = ql.SimpleQuote(100.0)
        helper = ql.FixedRateBondHelper(ql.QuoteHandle(price),sett,face,schedule,[row.ytc],ql.SimpleDayCounter())
        bond_helpers.append(helper)

    curve = ql.FittedBondDiscountCurve(0,cale,bond_helpers,ql.SimpleDayCounter(),ql.SvenssonFitting())
    curve.enableExtrapolation()
    
    tg = np.arange(0,30,1/12)
    tenors = []
    discountrates = []
    zerorates = []
    forwardrates = []
    ytms = []
    for yrs in tg:
        d = today + int(yrs*365)
        tenors.append(yrs)
        dr = curve.discount(yrs)
        discountrates.append(dr)
        zerorate = curve.zeroRate(yrs,compounding,ql.Semiannual)
        eqrate = zerorate.equivalentRate(dcnt,compounding,ql.Semiannual,today,d).rate()
        zerorates.append(eqrate*100)
        forwrate = curve.forwardRate(d,cale.advance(d,int(1),ql.Years),dcnt,compounding).rate()
        forwardrates.append(forwrate*100)
        d1 = cale.advance(today,0,ql.Days)
        ytm = curve.forwardRate(d1,d,dcnt,compounding).rate()
        ytms.append(ytm*100)
    
    ts = pd.DataFrame({'tenors':tenors,'discountrates':discountrates,'zerorates':zerorates,'forwardrates':forwardrates,
                  'yield2maturity':ytms})
    return ts

ts1 = yldcurvedata(tradda[tradda.type == 'green bond'])
print("1st finished")
ts1a = yldcurvedata(tradda[((tradda.rating=='A') | (tradda.rating=='AA') | (tradda.rating=='AAA')) & (tradda.type == 'green bond')])
print("2nd finished")
ts1b = yldcurvedata(tradda[((tradda.rating=='B') | (tradda.rating=='BB') | (tradda.rating=='BBB')) & (tradda.type == 'green bond')])
print("3rd finished")
ts1n = yldcurvedata(tradda[(tradda.rating=='NR') & (tradda.type == 'green bond')])
print("4th finished")
ts2 = yldcurvedata(tradda[tradda.type == 'brown bond'])
print("5th finished")
ts2a = yldcurvedata(tradda[((tradda.rating=='A') | (tradda.rating=='AA') | (tradda.rating=='AAA')) & (tradda.type == 'brown bond')])
print("6th finished")
ts2b = yldcurvedata(tradda[((tradda.rating=='B') | (tradda.rating=='BB') | (tradda.rating=='BBB')) & (tradda.type == 'brown bond')])
print("7th finished")
ts2n = yldcurvedata(tradda[(tradda.rating=='NR') & (tradda.type == 'brown bond')])
print("8th finished")
pickle.dump([ts1,ts1a,ts1b,ts1n,ts2,ts2a,ts2b,ts2n],open( "curves.p","wb" ))
```
